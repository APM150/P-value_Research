```python
import numpy as np
import mltools as ml
from Bagging import Bagging
import matplotlib.pyplot as plt
import time
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
import torch
```

## Iris Dataset

```python
iris = np.genfromtxt("data/iris.txt", delimiter=None)
Y = np.array([int(i) for i in iris[:, -1]])
X = iris[:, 0:-1]
np.random.seed(0)
X,Y = ml.shuffleData(X,Y)
Xtr, Xva, Ytr, Yva = ml.splitData(X, Y, 0.8)
```

```python
t0 = time.time()
model = Bagging()
model.train(Xtr, Ytr, percentageForClassifiers=[0.7, 0.15, 0.15], nBag=10, n_boot=1000)
print("Time:", time.time() - t0)
```

```python
# allDataPredict = allDataModel.all_p_values(Xva)
predictIris = model.all_p_values(Xva)
print(predictIris)
```

```python
print(model.score(Xva, Yva, predictIris))
```

```python
neigh0 = KNeighborsClassifier(n_neighbors=2)
neigh1 = KNeighborsClassifier(n_neighbors=2)
neigh2 = KNeighborsClassifier(n_neighbors=2)
neigh0.fit(Xtr[Ytr==0,:], Ytr[Ytr==0])
neigh1.fit(Xtr[Ytr==1,:], Ytr[Ytr==1])
neigh2.fit(Xtr[Ytr==2,:], Ytr[Ytr==2])
```

```python
# class 0
nearesetNeighborDistance0 = neigh0.kneighbors(Xva)[0][:, 0]
sortedNearestNeighborDistance0 = sorted(nearesetNeighborDistance0)
empiricalP_value0 = []
for i in range(len(nearesetNeighborDistance0)):
    empiricalP_value0.append(1 - ((sortedNearestNeighborDistance0.index(nearesetNeighborDistance0[i]) + 1) / (len(nearesetNeighborDistance0)+1)))

# class 1
nearesetNeighborDistance1 = neigh1.kneighbors(Xva)[0][:, 0]
sortedNearestNeighborDistance1 = sorted(nearesetNeighborDistance1)
empiricalP_value1 = []
for i in range(len(nearesetNeighborDistance1)):
    empiricalP_value1.append(1 - ((sortedNearestNeighborDistance1.index(nearesetNeighborDistance1[i]) + 1) / (len(nearesetNeighborDistance1)+1)))

# class 2
nearesetNeighborDistance2 = neigh2.kneighbors(Xva)[0][:, 0]
sortedNearestNeighborDistance2 = sorted(nearesetNeighborDistance2)
empiricalP_value2 = []
for i in range(len(nearesetNeighborDistance2)):
    empiricalP_value2.append(1 - ((sortedNearestNeighborDistance2.index(nearesetNeighborDistance2[i]) + 1) / (len(nearesetNeighborDistance2)+1)))

empiricalP_value = []
empiricalP_value.append(empiricalP_value0)
empiricalP_value.append(empiricalP_value1)
empiricalP_value.append(empiricalP_value2)
empiricalP_value = np.array(empiricalP_value)
empiricalP_value = empiricalP_value.T
print(empiricalP_value)
```

```python
colors = ['blue', 'green', 'red']
fig, ax = plt.subplots(1, 3, figsize=(20,5))
for i in range(3):
    ax[i].plot(np.linspace(1e-20, 1, 1000), np.linspace(1e-20, 1, 1000))
    for j in range(3):
        mask = Yva==j
        ax[i].scatter(empiricalP_value[mask,i], predictIris[mask,i], c=colors[j], label=f"true class {j}")
    ax[i].set_title(f"predict iris class {i} ([0.7, 0.15, 0.15], nBag=10, n_boot=1000)")
    ax[i].set_xlabel("Empirical p-value")
    ax[i].set_ylabel("Bagging predicred p-value")
    ax[i].legend()
plt.show()
```

## Wine Quality Dataset

```python
df = pd.read_csv('data/winequality-red.csv')
dx = df.sample(frac=1)
```

```python
X_s = np.array(dx.drop('quality', 1).values, dtype=np.float32)
y_s = np.array(dx['quality'].values - 3)
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X_s, y_s, test_size=0.2)
print(X_s.shape)
print(y_s.shape)
print(np.unique(y_test_s))
```

```python
t0 = time.time()
wineQualityModel = Bagging()
wineQualityModel.train(X_train_s, y_train_s, percentageForClassifiers=[0.1, 0.9, 0.0], nBag=10, n_boot=X_train_s.shape[0])
print("Time:", time.time() - t0)
```

```python
t0 = time.time()
predict = wineQualityModel.all_p_values(X_test_s)
print(predict)
print(time.time() - t0)
```

```python
print(wineQualityModel.score(X_test_s, y_test_s, predict))
```

```python
wineNeighs = []
for i in range(6):
    neigh = KNeighborsClassifier(n_neighbors=1)
    neigh.fit(X_train_s[y_train_s==i,:], y_train_s[y_train_s==i])
    wineNeighs.append(neigh)
```

```python
empiricalP_values = []
# class 0
for c in range(6):
    nearesetNeighborDistance = wineNeighs[c].kneighbors(X_test_s)[0][:, 0]
    sortedNearestNeighborDistance = sorted(nearesetNeighborDistance)
    empiricalP_value = []
    for i in range(len(nearesetNeighborDistance)):
        empiricalP_value.append(1 - ((sortedNearestNeighborDistance.index(nearesetNeighborDistance[i]) + 1) / (len(nearesetNeighborDistance)+1)))
    empiricalP_values.append(empiricalP_value)
empiricalP_values = np.array(empiricalP_values).T
print(empiricalP_values[y_test_s == 3, 3])
print(predict.shape)
```

```python
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']
fig, ax = plt.subplots(2, 3, figsize=(20,10))
for i in range(len(colors)):
    ax[i // 3][i % 3].plot(np.linspace(1e-20, 1, 1000), np.linspace(1e-20, 1, 1000))
    for j in range(len(colors)):
        mask = y_test_s==j
        ax[i // 3][i % 3].scatter(empiricalP_values[mask,i], predict[mask,i], c=colors[j], label=f"true class {j}")
    ax[i // 3][i % 3].set_title(f"predict wine class {i} ([0.7, 0.3, 0], nBag=10, n_boot=1000)")
    ax[i // 3][i % 3].set_xlabel("Empirical p-value")
    ax[i // 3][i % 3].set_ylabel("Bagging predicred p-value")
    ax[i // 3][i % 3].legend()
plt.show()
```

### Wine Dataset

```python
DATA=np.loadtxt("data/wine.data", delimiter=',')
X = DATA[:,1:]
Y = DATA[:,0]
Y = Y - 1
Y = np.array(Y, dtype=np.int)
DATA = np.concatenate((X, np.reshape(Y, (len(Y), 1))), axis=1)
np.random.shuffle(DATA)
training_portion = 0.8
index = int(len(DATA) * training_portion)
TRAIN = DATA[:index, :]
TEST  = DATA[index:, :]
Xtrain, Ytrain = TRAIN[:, :-1], TRAIN[:, -1]
Xtest , Ytest  = TEST[:, :-1], TEST[:, -1]
Ytrain = np.array(Ytrain, dtype=np.int)
Ytest = np.array(Ytest, dtype=np.int)
print(X.shape)
print(Y.shape)
print(np.unique(Ytrain))
print(Ytest.shape)
```

```python
t0 = time.time()
wineModel = Bagging()
wineModel.train(Xtrain, Ytrain, percentageForClassifiers=[0.0, 0.5, 0.5], nBag=10, n_boot=Xtrain.shape[0])
print("Time:", time.time() - t0)
```

```python
t0 = time.time()
predict = wineModel.all_p_values(Xtest)
print(predict)
print(time.time() - t0)
```

```python
print(wineModel.score(Xtest, Ytest, predict))
```
